## Running on Test Data + Generating Pearsons

```python
torch.cuda.empty_cache()  # Clear unused memory from GPU
torch.cuda.ipc_collect()
```

```python
from tqdm import tqdm
from scipy.stats import pearsonr
import torch
import numpy as np
from sklearn.preprocessing import StandardScaler


def compute_pearsons(model):

  # Initialize storage for predictions and targets for all tracks
  all_model_predictions = []
  all_model_retarded_predictions = []
  all_targets = []

  mapping = {'A': 0, 'C': 1, 'T': 2, 'G': 3, 'N': 4}
  required_sequence_length = 114688

  TEST_EXAMPLES = len(test_dataset)  # For testing, reduce to 10 examples

  for i in tqdm(range(TEST_EXAMPLES)):
      # Fetch the sequence and target data
      raw_sequence = test_dataset[i]['sequence']
      example_dna_sequence = torch.tensor([mapping[base] for base in raw_sequence], dtype=torch.long)

      # Truncate the sequence to the required length (if longer than expected)
      example_dna_sequence = example_dna_sequence[:required_sequence_length]
      example_dna_sequence = example_dna_sequence.unsqueeze(0)
      one_hot_sequence = seq_indices_to_one_hot(example_dna_sequence)

      # Fetch and process target data
      target = torch.tensor(test_dataset[i]['labels']).float()
      target = torch.log1p(target)

      one_hot_sequence = one_hot_sequence.cuda()
      target = target.cuda()

      # Compute outputs for the models
      output = model(one_hot_sequence)['human']


      # Store predictions and targets for all tracks
      all_model_predictions.append(output[0].detach().cpu().numpy())
      all_targets.append(target.detach().cpu().numpy())

    # Convert the collected data into numpy arrays
  all_model_predictions = np.concatenate(all_model_predictions, axis=0)  # Shape: [Total_bins, 50]
  all_targets = np.concatenate(all_targets, axis=0)  # Shape: [Total_bins, 50]
  # Compute Pearson correlations for each track
  model_pearson_correlations = []

  for track_idx in range(all_targets.shape[1]):  # Loop through 50 tracks
      track_predictions = all_model_predictions[:, track_idx]  # Shape: [Total_bins]
      track_target = all_targets[:, track_idx]

      r_value_model, _ = pearsonr(track_predictions, track_target)
      model_pearson_correlations.append(r_value_model)
  return model_pearson_correlations

model_pearson_correlations = compute_pearsons(model)
ae_pearson_correlations = compute_pearsons(model_ae)

# Optional: Print or analyze the correlations
print("Model Pearson correlations for each track:", model_pearson_correlations)
print("Model Pearson correlations for each track:", ae_pearson_correlations)
```

```python
# Pearson Plots
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming model_pearson_correlations and model_retarded_pearson_correlations are computed earlier

model_pearson_correlations = np.array(model_pearson_correlations)
#model_retarded_pearson_correlations = np.array(model_retarded_pearson_correlations)

# Compute the average R values for both models
average_r_model = np.mean(model_pearson_correlations)
average_r_model_retarded = np.mean(model_pearson_correlations)

print(f"Average R value for Model: {average_r_model:.3f}")
print(f"Average R value for Model Retarded: {average_r_model_retarded:.3f}")

# Create a scatter plot to compare the R values
plt.figure(figsize=(8, 8))
sns.scatterplot(
    x=model_pearson_correlations,
    y=model_pearson_correlations,
    color="black",
    s=10
)

# Add diagonal line (y=x) for reference
plt.plot([0, 1], [0, 1], color="gray", linestyle="--")

# Add text annotations for the average R values
plt.text(0.1, 0.9, f"{average_r_model:.3f}", fontsize=18, color="black")
plt.text(0.9, 0.1, f"{average_r_model_retarded:.3f}", fontsize=18, color="black")

plt.xlabel("Base Enformer R Values", fontsize=14)
plt.ylabel("Encoder-Enhanced Enformer R Values", fontsize=14)
plt.title("CAGE Pearson R Comparison", fontsize=16)

# Show the plot
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
